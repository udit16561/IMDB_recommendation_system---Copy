```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/data/movies.csv')
df.head(5)
```

```python
df = df.drop(['year','certificate', 'duration', 'votes', 'gross_income', 'directors_id', 'stars_id'], axis=1)
df.head(5)
```

```python
import numpy as np

# Create a NumPy array of floats
float_array = df['rating']

# Convert the float array to a string array
df['rating'] = float_array.astype(str)
```

```python
df['tags'] = df['genre'] + df['directors_name'] + df['stars_name'] + df['description']
```

```python
df['description'][0]
```

```python
dt = df.drop(columns=['genre','directors_name','stars_name','description'])
```

```python
dt.head()
```

```python
dt['tags'] = dt['tags'].apply(lambda x: "".join(x))
dt.head()
```

```python
dt=dt.iloc[:100000,:]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(dt['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
dt['tags']=dt['tags'].apply(stem)
```

```python
dt['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = df[df['name'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0],1])
```

```python
recommend('Tehran')
```
